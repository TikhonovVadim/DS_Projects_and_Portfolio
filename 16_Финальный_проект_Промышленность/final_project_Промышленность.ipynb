# Описание этапа обработки
Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша. 

Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (англ. tribe, «масса»).

Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. 

Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки. Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (англ. slab, «плита»).

**сформируем функции для повторяющихся действий/выводов по датасетам**

```python
# функция для вывода информации по датафрейму 
def func_data_info(data):
    """
    функция выводит информацию по датафрейму
    метод info() - общая информация
    метод isna().mean() - данные по количеству пропусков в % в разрере колонок/признаков
    метод dublicated() - количество явных дубликтов
    метод describe() - Описательная статистика по числовым признакам
    """   
    print('  Общая информация по датафрейму')
    print(data.info())
    print()
    print('  Данные по количеству пропусков в %')
    print(round(data.isna().mean(),2)*100)
    print()
    print(f'  Количество явных дубликтов - {data.duplicated().sum()}')
    print()
    print('  Описательная статистика по числовым признакам')
    print(data.describe()) 
```

```python
def func_make_histogram_or_boxplot(data, column=str, kind=str):
    """
    функция строит графики Гистрограм или Диаграммы размаха
    параметры:
    data - датасет с данными - dataframe
    column - колонка/ признак по который нужно построит график - str
    kind - тип графика: hist = Гистрограмма, box = Диаграмма размаха - str  
    """
    if kind=='hist':
        data[column].plot(kind='hist', bins=50, figsize=(14,4),
                    grid=True).set_title('Гистограмма распределения признака - ' + column, fontsize=16)
        plt.xlabel('Значения - ' + column, fontsize=12) 
        plt.ylabel('Частота')
        
    elif kind=='box':
        data[column].plot(kind='box', figsize=(14,4), vert=False, use_index=False, 
                                   grid=True).set_title('Диаграмма размаха признака - ' + column, fontsize=16)
        plt.xlabel('Значения - ' + column, fontsize=12);
        plt.ylabel('Медиана, Квартили, Выбросы')
        
    else:
        print('В функцию передан не поддерживаемый тип графика')
        
    plt.show()
```

### Формирование не менее 2-х моделей с подбором гиперпараметров на cross-validation (CV)

в данном проекте используем 4 разные модели, чтобы иметь возможность сравнения:
- Линейная регрессия с масштабированием признаков
- Ридж регрессия (линейная регрессия с регуляризацией)
- Случайный лес

**Линейная регресия**

**масштабируем признаки для LinearRegression() методом StandardScaler**

```python
# инициализируем и обучим scaler 
scaler = StandardScaler()
scaler.fit(X_train) 
```

```python
# преобразуем выборки функцией transform()
X_train_st = scaler.transform(X_train)
pd.options.mode.chained_assignment = None
```

```python
# GridSearchCV для модели LinearRegression() 

params = {}

grid_LR = GridSearchCV(LinearRegression(), param_grid=params, cv=3, scoring='neg_mean_absolute_error', n_jobs=-1)
```

```python
%%time

# подбор параметров на fit
start_time = time.time()
grid_LR_model = grid_LR.fit(X_train_st, y_train)

end_time = time.time()
training_time_lr = end_time - start_time
```

```python
# вывод лучшей оценки
mae_lr = round(abs(grid_LR_model.best_score_),2)
mae_lr
```

**Ридж регрессия**

```python
# GridSearchCV для модели RidgeRegression() 

params = {'alpha':[10**x for x in range(-4,4)]}

grid_Ridge = GridSearchCV(Ridge(), 
                          param_grid=params, 
                          cv=3, 
                          scoring='neg_mean_absolute_error', 
                          n_jobs=-1)
```

```python
%%time

# подбор параметров на fit
start_time = time.time()
grid_Ridge_model = grid_Ridge.fit(X_train_st, y_train)

end_time = time.time()
training_time_ridge = end_time - start_time
```

```python
# вывод лучшей оценки
mae_ridge = round(abs(grid_Ridge_model.best_score_),2)
mae_ridge
```

```python
# вывод параметров лучшей модели
grid_Ridge_model.best_params_
```

**Случайный лес**

```python
# GridSearchCV для модели RandomForestRegressor()

params = { 
    'n_estimators': [50, 100, 200, 500],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [3,4,5,6,7,8,9,10]
}

grid_RFR = GridSearchCV(RandomForestRegressor(random_state=RANDOM_STATE), 
                        param_grid=params, 
                        scoring='neg_mean_absolute_error',  
                        cv=3,
                        n_jobs=-1)
```
